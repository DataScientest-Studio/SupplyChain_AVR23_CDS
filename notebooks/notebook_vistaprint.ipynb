```python
df_vp=pd.read_csv('VistaPrint.csv')
df_vp.head()
```

```python
df_vp.info()
```

```python
# verification des doublons:

n=df_vp.duplicated().sum()
print("il y a",n,"doublons dans la DataFrame df_vp ")

# suppression des doublons:
df_vp.drop_duplicates(keep='first')
```

```python
# Renommer les colonnes
dic={'Like':'Client_name','Title_URL':'URL_Source', 'typography_bodym__xgxz_': 'Location',
      'Image':'Rate','typography_bodym__xgxz_1':'comment_date','Like2':'Title_comment',
     'typography_bodyl__kuyfj':'Comment','typography_bodym__xgxz_3':'date_experience',
     'typography_bodym__xgxz_7':'Reply','Reply6':'Reply_date'}

df_vp=df_vp.rename(dic,axis=1)

# supprimer la colonne URL_Source
df_vp = df_vp.drop('URL_Source', axis=1)

df_vp.head()
```

```python
# modification de la colonne Rate et date_experience :
df_vp['Rate'] = df_vp['Rate'].str.replace('https://cdn.trustpilot.net/brand-assets/4.1.0/stars/stars-', '')
df_vp['Rate'] = df_vp['Rate'].str.replace('.svg', '')
df_vp['date_experience'] = df_vp['date_experience'].str.replace("Date de l'expérience: ", '')

df_vp.head()
```

```python
# Création d'un nouveau DataFrame sans NaN:
df_vp_pur=df_vp.dropna(axis = 0, how = 'any', subset = ['Reply','Reply_date'])
# df_vp_pur sera utilisé pour la visualisation des graphiques 
# incluant reply et replay_date
```
